https://www.acmicpc.net/problem/11403


```python
N = int(input().rstrip())
m = [list(map(int,input().rstrip().split())) for _ in range(N)]

for k in range(0, N):         # 경유지 노드
    for i in range(0, N):     # 출발 노드
        for j in range(0, N): # 도착 노드
            if m[i][k]==1 and m[k][j]==1:
                m[i][j] = 1
for t in m:
    print(*t)
```
