```python
import sys

n = int(input())

time = [tuple(map(int,input().split())) for _ in range(n)]

print(time)

time = sorted(time, key = lambda x : (x[1],x[0]))

print(time)

cnt = 0
end = 0

for i in time:
    if i[0] >= end:
        print('i : ',i,'end : ',end)
        cnt += 1
        end = i[1]

print(cnt)
```
