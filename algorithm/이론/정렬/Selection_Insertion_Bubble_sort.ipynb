# Selection Sort (선택정렬) - 안정정렬

### 1. 왼쪽부터 시작해서 가장 작은 노드와 swap

### 2. swap이 이루어진 노드를 제외한 처음 노드( 왼쪽 부터 2번째 )와 가장 작은 노드를 다시 swap 

### 3. 위 과정을 반복한다.

## 시간복잡도 O(N^2)

### 외부루프 : (N-1)
### 내부루프 : (N-1),(N-2),...,2,1
### 교환횟수 : 3(n-1)


# Python Code

```python
def selection_sort(lst):
    for i in range(len(lst) - 1): # 인덱스 이용을 위함. 0부터 시작이므로 길이에서 1을 빼줌
        min_idx = i               # 처음 최솟값은 lst[0]으로 지정하고 이를 비교해가면서 갱신시킴.  
        for j in range(i+1,len(lst)): # 시작부분은 정해졌고, 시작 부분을 제외한 나머지원소들에 대해서
            if lst[min_idx] > lst[j]: # 시작부분(최솟값 이라고 정한 것)과 j 번째 원소를 비교해서 j번째 가 더 작다면
                min_idx = j           # 최솟값을 바꿔줌 (해당 값은 인덱스)
        
        lst[i],lst[min_idx] = lst[min_idx],lst[i]
        # for 구문 내부에서 둘의 위치를 바꿔주는게 아니라 외부에서 변경
    return lst
```

# Test

```python
test_lst = [5,0,10,22,34,1,9,3] # input

selection_sort(test_lst)
```

# Insertion Sort (삽입정렬) - 불안정정렬


### 1. 값이 낮아지는 곳의 index를 찾는다.
### 2. 해당 lst[index]보다 큰 값을 좌측에서 찾는다.
### 3. 해당 큰 값을 갖는 index와 바꾼다.

## 시간복잡도

### 리스트가 이미 정렬된 경우 : O(N)
### 내림차순으로 정렬된 리스트를 오름차순으로 정렬하는 경우 : O(N^2)

# Python Code

```python
def ins_sort(lst):
    for end in range(1,len(lst)):
        insert = lst[end]
        i = end
        while i >0 and lst[i-1] > insert:
            lst[i] = lst[i-1]
            i -= 1
        
        lst[i] = insert
    
    return lst
```

# Test

```python
test =[5,3,1,6,8,9,7]
ins_sort(test)
```

# Bubble Sort (버블정렬) -안정정렬


### 1. index기준 0부터 시작해서 시작부분과 그 옆 부분만을 비교 후 위치를 바꾼다.
### 2. 다음 index를 시작지점으로 잡아 같은 행위를 반복한다.

### ( 해당 반복이 한 번 끝나면 가장 큰 값은 마지막 인덱스에 위치된다.)

## 시간복잡도
### 내부.외부 루프가 전부 n 번씩 반복하므로 O(N^2)

# Python Code

```python
def bubble_sort(lst):
    end = len(lst) - 1
    while end > 0:
        finish_idx = 0
        for i in range(end):
            if lst[i] > lst[i+1]:
                lst[i],lst[i+1] = lst[i+1],lst[i]
                finish_idx = i
        end = finish_idx
    return lst
```

# Test

```python
test = [10,2,1,7,4,3,9]
bubble_sort(test)
```
