## Stack (LIFO)


```python
stack = []
# append / pop  O(1)
stack.append(5)
stack.append(2)
stack.append(3)
stack.append(7)
print(stack)
stack.pop()
print(stack)
stack.append(1)
stack.append(4)
stack.pop()

print(stack[::-1])
print(stack)
```

## Queue (FIFO)


```python
from collections import deque

queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
print(queue)
queue.popleft()
print(queue)
queue.append(1)
queue.append(4)

print(queue)
queue.reverse() #     .
print(queue)
```

```python
def recursive_func(c):
    print('  ')
    if c == 10:
        return
    else:
        recursive_func(c+1)

recursive_func(0)
```

```python
# 

def dfs(g,v,visted):
    visited[v] = True
    print(v, end=' ')
    for i in g[v]:
        if not visited[i]:
            dfs(g,i,visited)

g = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visited = [False] * 9

dfs(g,1,visited)
```

```python
# 
def dfs_recur(node, dfs_graph, dfs_list=[]):
    stack =[]
    stack.append(node)
    
    while stack:
        node = stack.pop()
        
        if node not in dfs_list:
            dfs_list.append(node)
            dfs_graph[node].sort(reverse = True)
            stack.extend(dfs_graph[node])
    return dfs_list

dfs_graph = {
    1: [2,3,4],
    2: [5],
    3: [6],
    4: [],
    5: [7],
    6: [5],
    7: [6],
}

dfs_recur(2,dfs_graph,dfs_list=[])
```

```python
check = [False for _ in range(len(dfs_graph)+1)]
         
def dfs_recur(node, dfs_graph, dfs_list=[]):
    check[node] = True
    dfs_list.append(node)
    for i in dfs_graph[node]:
         if check[i] == False:
             dfs_recur(i,dfs_graph,dfs_list)
    return dfs_list

dfs_graph = {
    1: [2,3,4],
    2: [5],
    3: [6],
    4: [],
    5: [7],
    6: [5],
    7: [6],
}

dfs_recur(2,dfs_graph,dfs_list=[])
```

```python
def dfs_recur(node,g,dfs_list):
    dfs_list.append(node)
    
    for i in g[node]:
        #    ()
        if i ==[]:
            break
        #      
        if i not in dfs_list:
            dfs_recur(i,g,dfs_list)
        return dfs_list

g = {
1: [2,3,4],
2: [5],
3: [6],
4: [],
5: [7],
6: [5],
7: [6],
}

dfs_recur(2,dfs_graph,dfs_list=[])
```

## BFS (  )
### 1.       
### 2.                
### 3.   2     

```python
from collections import deque

def bfs(g,start,visited):
    queue = deque([start])
    visited[start] = True
    while queue:
        v = queue.popleft()
        print(v, end =' ')
        
        for i in g[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
g = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visited = [False] * 9

bfs(g,1,visited)
```

```python
def bfs_queue(start_node, bfs_graph):
    from collections import deque
    bfs_list=[]
    queue = deque([start_node])
    bfs_list.append(start_node)
    
    while queue:
        popped = queue.popleft()
        
        for i in bfs_graph[popped]:
            if i not in bfs_list:
                queue.append(i)
                bfs_list.append(i)
    return bfs_list

bfs_graph = { 
    1: [2,3,4],
    2: [5],
    3: [6],
    4: [],
    5: [7],
    6: [5],
    7: [6],
}

bfs_queue(2,bfs_graph)
```
