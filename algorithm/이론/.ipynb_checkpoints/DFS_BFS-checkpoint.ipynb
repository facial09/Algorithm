## Stack (LIFO)


```python
stack = []
# append / pop 모두 O(1)
stack.append(5)
stack.append(2)
stack.append(3)
stack.append(7)
print(stack)
stack.pop()
print(stack)
stack.append(1)
stack.append(4)
stack.pop()

print(stack[::-1])
print(stack)
```

## Queue (FIFO)


```python
from collections import deque

queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
print(queue)
queue.popleft()
print(queue)
queue.append(1)
queue.append(4)

print(queue)
queue.reverse() # 나중에 들어온 순서로 바꾸기 위함.
print(queue)
```

## 재귀함수 (Recursive Function)

```python
def recursive_func(c):
    print('재귀 함수를 호출합니다')
    if c == 10:
        return
    else:
        recursive_func(c+1)

recursive_func(0)
```

## DFS ( 깊이 우선 탐색 )
### 1. 탐색 시작 노드를 스택에 넣고 방문 처리
### 2. 스택의 최상단 노드에 방문하지 않은 인접한 노드가 하나라도 있으면 그 노드를 스택에 넣고 방문 처리한다. 방문하지 않은 인접 노드가 없으면 스택에서 최상단 노드를 꺼낸다.
### 3. 더 이상 2번의 과정을 수행할 수 없을 때까지 반복

```python
# 재귀함수

def dfs(g,v,visted):
    visited[v] = True
    print(v, end=' ')
    for i in g[v]:
        if not visited[i]:
            dfs(g,i,visited)

g = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visited = [False] * 9

dfs(g,1,visited)
```

```python
# 반복문
def dfs_recur(node, dfs_graph, dfs_list=[]):
    stack =[]
    stack.append(node)
    
    while stack:
        node = stack.pop()
        
        if node not in dfs_list:
            dfs_list.append(node)
            dfs_graph[node].sort(reverse = True)
            stack.extend(dfs_graph[node])
    return dfs_list

dfs_graph = {
    1: [2,3,4],
    2: [5],
    3: [6],
    4: [],
    5: [7],
    6: [5],
    7: [6],
}

dfs_recur(2,dfs_graph,dfs_list=[])
```

```python
check = [False for _ in range(len(dfs_graph)+1)]
         
def dfs_recur(node, dfs_graph, dfs_list=[]):
    check[node] = True
    dfs_list.append(node)
    for i in dfs_graph[node]:
         if check[i] == False:
             dfs_recur(i,dfs_graph,dfs_list)
    return dfs_list

dfs_graph = {
    1: [2,3,4],
    2: [5],
    3: [6],
    4: [],
    5: [7],
    6: [5],
    7: [6],
}

dfs_recur(2,dfs_graph,dfs_list=[])
```

```python
def dfs_recur(node,g,dfs_list):
    dfs_list.append(node)
    
    for i in g[node]:
        # 빈 노드의 경우 (종료조건)
        if i ==[]:
            break
        # 방문하지 않은 노드   
        if i not in dfs_list:
            dfs_recur(i,g,dfs_list)
        return dfs_list

g = {
1: [2,3,4],
2: [5],
3: [6],
4: [],
5: [7],
6: [5],
7: [6],
}

dfs_recur(2,dfs_graph,dfs_list=[])
```

## BFS ( 너비우선탐색 )
### 1. 탐색 시작 노드를 큐에 삽입하고 방문 처리
### 2. 큐에서 노드를 꺼낸 뒤 해당 노드의 인접 노드 중에서 방문하지 않은 노드를 모두 큐에 삽입하고 방문처리
### 3. 더 이상 2번 과정 수행 불가 까지 반복

```python
from collections import deque

def bfs(g,start,visited):
    queue = deque([start])
    visited[start] = True
    while queue:
        v = queue.popleft()
        print(v, end =' ')
        
        for i in g[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
g = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

visited = [False] * 9

bfs(g,1,visited)
```

```python
def bfs_queue(start_node, bfs_graph):
    from collections import deque
    bfs_list=[]
    queue = deque([start_node])
    bfs_list.append(start_node)
    
    while queue:
        popped = queue.popleft()
        
        for i in bfs_graph[popped]:
            if i not in bfs_list:
                queue.append(i)
                bfs_list.append(i)
    return bfs_list

bfs_graph = { 
    1: [2,3,4],
    2: [5],
    3: [6],
    4: [],
    5: [7],
    6: [5],
    7: [6],
}

bfs_queue(2,bfs_graph)
```
